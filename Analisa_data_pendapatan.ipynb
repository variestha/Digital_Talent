# 1. Deskripsi Data
Data ini berisi mengenai nilai ekspor dan impor negara Indonesia.
Penjelasan mengenai variabel pada data: 
a. tahun: Tahun 
-> data diambil dari tahun 1990 sampai tahun 2013. 

b. Sumber_Pendapatan: Pajak / Bukan Pajak / Hibah
-> dari tahun 1990 sampai tahun 2013. 

c. Keterangan_sumber_pendapatan = menerangkan sumber pendapatan
-> SDA, Gas Alam, PBLU, SBI, Hibah , PPN, BPTHB, Laba BUMN dan lain-lain

d. nilai =Nilai dari Pos Pendapatan tersebut (dalam Milyar Rupiah)

Catatan:

1. Sampai dengan tahun 2004, realisasi menggunakan angka Perhitungan Anggaran Negara (PAN).

2. Sejak tahun 2005, menggunakan angka Laporan Keuangan Pemerintah Pusat (LKPP).

3. Sejak Januari 2012, data sudah disesuaikan dengan publikasi resmi KemenKeu.

```python
s = data_apbn.nilai.sum()
mi = data_apbn.nilai.min()
ma = data_apbn.nilai.max()
me = data_apbn.nilai.mean()
md = data_apbn.nilai.median()


print('==================================================')
print('Jumlah nilai mata uang adalah',s)
print('==================================================')
print('Nilai mata uang minimum',mi)
print('==================================================')
print('Nilai mata uang maksimum',ma)
print('==================================================')
print('Rata-rata dari nilai mata uang adalah',me)
print('==================================================')
print('Median dari nilai mata uang adalah',md)
```

```python
ax = data_apbn.groupby('tahun').mean()['nilai']
ax = ax.plot(kind='line', title ="Rata-Rata Nilai Mata Uang Berdasarkan Tahun", figsize=(10,5), legend=True, fontsize=12)
ax.set_xlabel("Tahun", fontsize=12)
ax.set_ylabel("Nilai Mata Uang", fontsize=12)
```

```python
ax = data_apbn.groupby('tahun').mean()['nilai']
ax = ax.plot(kind='bar', title ="Rata-Rata Nilai Mata Uang Berdasarkan Tahun", figsize=(10,5), legend=True, fontsize=12)
ax.set_xlabel("Tahun", fontsize=12)
ax.set_ylabel("Nilai Mata Uang", fontsize=12)
```

```python
ax = data_apbn.groupby('tahun').mean()['nilai']
ax = ax.plot(kind='pie', title ="Rata-Rata Nilai Mata Uang Berdasarkan Tahun", figsize=(10,5), legend=True, fontsize=12)
ax.set_xlabel("Tahun", fontsize=12)
ax.set_ylabel("Nilai Mata Uang", fontsize=12)
```
